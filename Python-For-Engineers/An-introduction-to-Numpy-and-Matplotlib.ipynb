# 列表推导

```python
x = [5,10,15,20,25]
y = []

for counter in x:
    y.append(counter/5)
    

print("\nOld fashioned way: x = {} y = {} \n".format(x, y))
```

```python
z = [counter*5 for counter in x]
print(z)
```

实际上，我们可以通过 numpy 很容易的实现：

```python
import numpy as np

a = np.array(x)
b = a / 5
print(b)
```

# 绘制正弦、余弦波

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "-g", label="sine")
plt.plot(x, y2, "-b", label="cos")

plt.legend(loc="upper right")
plt.ylim(-1.5, 2)

plt.show()
```

# 绘制员工姓名及其薪水

```python
import numpy as np
import matplotlib.pyplot as plt

salary = np.fromfile("salaries.txt", dtype=int, sep=",")
names = np.genfromtxt("names.txt", dtype='str', delimiter=",")

print(salary)
```

```python
print(names)
```

```python
x = np.arange(len(names))

print(x)
```

```python
plt.bar(x, salary)
plt.xticks(x, names)

plt.ylabel("Salaries")
plt.xlabel("Names")
plt.title("Salary of 10 random people")

plt.show()
```

```python
print(np.max(salary), np.min(salary), np.average(salary), np.median(salary))
```
